```python
# Set up environment variables
import os
import sys

# 1. Set JAVA_HOME (Open JDK 11)
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-11-openjdk-amd64'

# 2. Set SPARK_HOME explicitly to the venv installation
# This is crucial so PySpark finds its own jars
import pyspark
os.environ['SPARK_HOME'] = pyspark.__path__[0]

# 3. Add to PATH
os.environ['PATH'] = f"{os.environ['JAVA_HOME']}/bin:{os.environ['SPARK_HOME']}/bin:{os.environ.get('PATH', '')}"

# 4. Set PySpark Python configuration (ensure it uses the current venv python)
os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable

print(f"JAVA_HOME: {os.environ['JAVA_HOME']}")
print(f"SPARK_HOME: {os.environ['SPARK_HOME']}")
print("Environment configured.")
```

# Exercises to Learn PySpark
This notebook contains exercises to help you learn PySpark fundamentals operations. Follow the instructions in each cell to complete the tasks.

```python
# Import necessary libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, avg, sum, max, min, when, lit
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType
```

```python
# Initialize Spark Session
spark = SparkSession.builder \
    .appName("PySpark Exercises") \
    .master("local[*]") \
    .getOrCreate()

print("Spark Session Created Successfully!")
print(f"Spark Version: {spark.version}")
```

Let's create a simple Spark DataFrame with some student sample data (name, age, grade, subject) and perform some basic operations.

```python
# Example data:
data = [
    ("Alice", 20, 85.5, "Math"),
    ("Bob", 21, 78.0, "Math"),
    ("Charlie", 19, 92.0, "Science"),
    ("David", 22, 88.5, "Science"),
    ("Eve", 20, 95.0, "Math")
]

schema = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("grade", DoubleType(), True),
    StructField("subject", StringType(), True)
])
```

## Exercise 1: Create a sample DataFrame 
Create a Spark DataFrame with the sample data provided and display its contents.

```python
df_students = spark.createDataFrame(data, schema=schema)
```

```python
# Display the DataFrame
print("\n=== Student Data ===")
df_students.show()
```

## Exercise 2: Basic operations - Filtering
Filter students with grade > 85 and display the results.

```python
df_students.filter(col("grade") > 85).show()
```

## Exercise 3: Basic operations - Aggregation I
Count students by age and display the results.

```python
df_students.groupBy("age").count().show()
```

## Exercise 4: Basic operations - Aggregation II
Calculate average grade by subject and display the results.

```python
df_students.groupBy("subject").agg(avg("grade").alias("average_grade")).show()
```
