## 1. Creating Your First Polars DataFrame

```python
import polars as pl
import random

# ----------------------------
# Create full synthetic dataset
# ----------------------------
N = 100_000

cities = ["Valencia", "Madrid", "Sevilla", "Bilbao", "Barcelona"]
categories = ["Electronics", "Fashion", "Home", "Sports", "Toys"]

df = pl.DataFrame({
    "order_id": list(range(1, N + 1)),
    "user_id": [random.randint(1, 1200) for _ in range(N)],
    "value": [random.randint(5, 10000) for _ in range(N)],
    "units": [random.randint(1, 5) for _ in range(N)],
    "city": [random.choice(cities) for _ in range(N)],
    "category": [random.choice(categories) for _ in range(N)],
    "is_active": [random.choice([True, False]) for _ in range(N)],
    "has_discount": [random.choice([True, False]) for _ in range(N)],
})
```

```python
# Task 1: direct selection
print(df.select(["order_id", "city", "value", "units"]).head())

# Task 2: select numeric columns
print(df.select(cs.numeric()).head())
```

### Exercise â€” Filtering high-value orders

Find all orders with:
- value > 9000

Then show only: city, value, category

ğŸ’¡ *Hint:* combine `.filter()` + `.select()`


```python
# Find all orders with value > 9000, show only city, value, category
print(df.filter(pl.col("value") > 9000).select(["city", "value", "category"]).head())
```

### Exercise â€” Classifying orders by size

Create a new column `size_label`:

- "large"  â†’ value > 7000  
- "medium" â†’ 3000 < value â‰¤ 7000  
- "small"  â†’ otherwise  

ğŸ’¡ Hint: use `pl.when().then().otherwise()`


```python
# Create size_label column
df_with_size = df.with_columns(
    pl.when(pl.col("value") > 7000).then(pl.lit("large"))
    .when(pl.col("value") > 3000).then(pl.lit("medium"))
    .otherwise(pl.lit("small"))
    .alias("size_label")
)
print(df_with_size.select(["value", "size_label"]).head(10))
```

### Mini-exercise â€” Pipeline

Tasks:
1. Select city, value, units
2. Filter units â‰¥ 3
3. Add value_per_unit = value / units


```python
# Pipeline: Select, Filter, Add column
pipeline_result = (
    df.select(["city", "value", "units"])
    .filter(pl.col("units") >= 3)
    .with_columns((pl.col("value") / pl.col("units")).alias("value_per_unit"))
)
print(pipeline_result.head())
```

### Mini-exercise â€” Category performance

1. Average value per category
2. Sort descending
3. Top 5 categories
4. Add an insight


```python
top_categories = (
    df.group_by("category")
    .agg(pl.col("value").mean().alias("avg_value"))
    .sort("avg_value", descending=True)
    .head(5)
)
print(top_categories)
# Insight: The average order value is relatively consistent across categories due to random generation.
```

### Exercise â€” City performance

Compute per city:
- number of orders
- average value
- total units sold


```python
result = df.group_by("city").agg([
    pl.len().alias("num_orders"),
    pl.col("value").mean().alias("avg_value"),
    pl.col("units").sum().alias("total_units")
])
print(result)
```

### Mini-exercise â€” Active vs inactive users

Compare active vs inactive users:

```python
# Compare active vs inactive users
compare_users = df.group_by("is_active").agg([
    pl.len().alias("num_orders"),
    pl.col("value").mean().alias("avg_order_value"),
    pl.col("value").max().alias("max_order_value")
])
print(compare_users)
```

## Exercise â€” Boolean logic & flags

Create a new column `is_high_value_active` that is:

- `true` if:
  - `value > 8000`
  - and `is_active = true`
- `false` otherwise

Return a sample of 10 rows showing:
- value
- is_active
- is_high_value_active


```python
# Boolean logic & flags
high_value_active = df.with_columns(
    ((pl.col("value") > 8000) & (pl.col("is_active") == True)).alias("is_high_value_active")
).select(["value", "is_active", "is_high_value_active"]).head(10)
print(high_value_active)
```

## Mini-exercise â€” Multiple filters

Find orders that:

- are from Madrid or Barcelona
- have units â‰¥ 2
- have value < 5000

Then show:
- city
- units
- value


```python
# Multiple filters solution
madrid_barcelona_filtered = df.filter(
    pl.col("city").is_in(["Madrid", "Barcelona"]),
    pl.col("units") >= 2,
    pl.col("value") < 5000
).select(["city", "units", "value"])
print(madrid_barcelona_filtered.head(10))
```

## Exercise â€” Derived metrics

Create two new columns:

- `value_per_unit` â†’ value / units
- `is_bulk_order` â†’ units â‰¥ 4

Return a sample of 10 rows.


```python
# Derived metrics solution
derived_df = df.with_columns([
    (pl.col("value") / pl.col("units")).alias("value_per_unit"),
    (pl.col("units") >= 4).alias("is_bulk_order")
])
print(derived_df.select(["value", "units", "value_per_unit", "is_bulk_order"]).head(10))
```

## Mini-exercise â€” Ranking (basic)

Rank orders within each city by value:

- highest value â†’ rank 1

Show:
- city
- value
- rank in city


```python
# Ranking solution
ranked_df = df.with_columns(
    pl.col("value").rank(descending=True).over("city").alias("rank_in_city")
)
print(ranked_df.select(["city", "value", "rank_in_city"]).head(10))
```

## Exercise â€” City summary table

Build a summary table per city with:

- number of orders
- total value
- average value
- maximum value

Sort results by total value descending.

Note:
Each aggregation must have a unique output column name.


```python
# City summary table solution
city_summary = df.group_by("city").agg([
    pl.len().alias("num_orders"),
    pl.col("value").sum().alias("total_value"),
    pl.col("value").mean().alias("avg_value"),
    pl.col("value").max().alias("max_value")
]).sort("total_value", descending=True)
print(city_summary)
```
